```python
df=spark.read.format('csv').option('header','true').load('/Volumes/workspace/databricks_poc/poc')
df.display()
```

```python
df = df.withColumnRenamed('Customer Id','customer_id') \
    .withColumnRenamed('First Name','first_name') \
    .withColumnRenamed('Last Name','last_name') \
    .withColumnRenamed('Phone 1','phone_1') \
    .withColumnRenamed('Phone 2','phone_2') \
    .withColumnRenamed('Subscription Date','subscription_date')
```

```python
df.display()
```

```python
df.write.format('delta').mode('overwrite').saveAsTable('cust_data')
```

```python
%sql
select * from cust_data
```
